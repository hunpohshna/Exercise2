## Create a Histogram for iris dataset

```python
import seaborn as sns

import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
# customize the histogram

sns.histplot(data=iris, x="petal_length", bins=20, color="green")
# add labels and title
sns.set_theme(style="whitegrid")
plt.xlabel("Petal Length (cm)")

plt.ylabel("Frequency")

plt.title("Distribution of Petal Lengths in Iris Flowers")
# display the plot
plt.show()
```

## Create a Scatter Plot for tips dataset

```python
import seaborn as sns
from matplotlib import pyplot as plt
# Load the Tips dataset
tips = sns.load_dataset("tips")
print(type(tips))
print(tips)
```

```python
sns.set_theme(style="whitegrid")
sns.scatterplot(x="total_bill", y="tip", data=tips,hue="smoker")
# add labels and title

plt.xlabel("Total Bill")

plt.ylabel("Tip")

plt.title("Relationship between Total Bill and Tip")
plt.show()
```

## Create a bar plot for titanic dataset

```python
titanic = sns.load_dataset("titanic")
print(titanic.head())
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x="alive", y="fare", hue="sex", color="yellow",palette='rocket', data=titanic)

plt.xlabel("Class")

plt.ylabel("Fare")
sns.set_style("whitegrid")
plt.title("Average Fare by Class and Gender on the Titanic")

# display the plot
plt.figure(figsize=(8, 6))
plt.show()
```

## create a Heatmap for tip and flight dataset

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load the dataset

tips = sns.load_dataset('tips')
# Create a heatmap of the correlation between variables
newdf=tips.iloc[:,[0,1,6]]
print(newdf)
corr = newdf.corr()
print(corr)
```

```python
sns.heatmap(corr,cmap='viridis',cbar=True, annot=True)
# Show the plot

plt.show()
```

```python
# Load the dataset

flights = sns.load_dataset('flights')
print(flights)
# Pivot the data

newdf = flights.pivot(index='month', columns='year',values= 'passengers')
print(newdf)
# Create a heatmap

```

```python
sns.heatmap(newdf, cmap='viridis', annot=True,cbar=True, fmt='d')

# Set the title and axis labels

plt.title('Passengers per month')

plt.xlabel('Year')

plt.ylabel('Month')

# Show the plot

plt.show()
```

## Create a Pairplot for Iris dataset

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Load iris dataset

iris = sns.load_dataset("iris")
print(iris)
```

```python
# Create pair plot with custom settings

sns.set_theme(style="whitegrid")
sns.pairplot(data=iris,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "rocket")
# Set title
plt.title("Iris Dataset Pair Plot")
# Show plot
#plt.figure(figsize=(8, 6))
plt.savefig(fname='figurepairplot',orientation='landscape')
plt.show()
```

## create a count plot for tips and titanic dataset

```python
 # Show value counts for two categorical variables and using hue parameter:

#While the points are plotted in two dimensions, another dimension can be added to the plot by coloring the points according to a third variable.
 
import seaborn as sns
import matplotlib.pyplot as plt

# read a tips.csv file from seaborn library
df = sns.load_dataset('tips')
sns.set_theme(style="whitegrid") 
# count plot on two categorical variable
sns.countplot(x ='sex', hue = "smoker", data = df)
 
# Show the plot
plt.show()
```

```python
#The count plot can be thought of as a histogram across a categorical variable.The example below demonstrates the count plot
import matplotlib.pyplot as plt
import seaborn as sns
 
# load dataset
titanic = sns.load_dataset('titanic')
# create plot
print(titanic.head())
 
sns.set_theme(style="whitegrid")
sns.countplot(x = 'class', hue = 'who', data = titanic, palette = 'magma')
plt.title('Category of passenger')
plt.show()
```
